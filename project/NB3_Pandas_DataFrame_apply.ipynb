# Using Pandas DataFrame.apply()
### A little exercise creating nominal attribute from numeric using Pandas DataFrame.apply()
* The exercise uses ".apply" (Pandas DataFrame) to create a nominal (categorical) attribute from a numeric attribute
* i.e. create 'buckets'
* Essentially what is done when creating a 'histogram'

```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
```

```python
# get the data
train_df = pd.read_csv('data/titanic/train.csv')
test_df = pd.read_csv('data/titanic/test.csv')
combine = [train_df, test_df]
```

```python
# create the Age_Bracket column (initializing to -1)
train_df["Age_Bracket"] = -1
```

```python
train_df.head(3)
```

```python
# use ".apply" to create a column by mapping from another column
# Mapping age

def age_to_age_bracket(cols):
    Age = cols[0]

    if Age <= 16: return 0
    if Age <= 32: return 1
    if Age <= 48: return 2
    if Age <= 64: return 3
    return 4
    
```

```python
# Updating the Age_Bracket column usingthe age_to_age_bracket function

train_df['Age_Bracket'] = train_df[['Age']].apply(age_to_age_bracket, axis = 1)
```
